# Ducting
***

Referring to Section 2.7.4, an atmospheric duct is a region in the lower layers of the atmosphere where the vertical gradient of the refractive index is such that the propagating waves are ducted or guided and tend to follow the curvature of the earth, as shown in Figure 2.10.  Ducts provide a mechanism for waves to propagate beyond the line of sight and limit the spread of the wavefront to only the horizontal dimension.  A duct is possible whenever the vertical refractivity gradient, $(dN / dh)$, is less than about $-157$ N/km.
***

Begin by getting library path

```python
import lib_path
```

Set up the duct thickness

```python
dts = [10.0, 20.0, 50.0]
```

Set up the refractivity gradient using the `linspace` routine from `scipy`

```python
from scipy import linspace

refractivity_gradient = linspace(-500., -150., 1000)
```

Set the line sytles for plotting the results and using the `matplotlib` routines

```python
from matplotlib import pyplot as plt

line_style = ['-', '--', '-.', ':', '-', '--', '-.', ':']
```

Calculate the critical angle for ducting

```python
from Libs.wave_propagation import ducting


# Set the figure size

plt.rcParams["figure.figsize"] = (15, 10)


i = 0

for dt in dts:

    critical_angle = ducting.critical_angle(refractivity_gradient, dt)

    i += 1



    # Display the results

    plt.plot(refractivity_gradient, critical_angle * 1e3, line_style[i-1], label="Thickness {} (m)".format(dt))


    
    # Set the plot title and labels

    plt.title('Ducting over a Spherical Earth', size=14)

    plt.xlabel('Refractivity Gradient (N/km)', size=12)

    plt.ylabel('Critical Angle (mrad)', size=12)



    # Set the tick label size

    plt.tick_params(labelsize=12)



    # Set the legend

    plt.legend(loc='best', prop={'size': 10})



    # Turn on the grid

    plt.grid(linestyle=':', linewidth=0.5)
```
