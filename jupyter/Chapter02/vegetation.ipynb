# Vegetation Attenuation

For some radar applications, the attenuation in vegetation can be important. However, there is a wide range of conditions and foliage types making it extremely difficult to develop generalized prediction methods.  Furthermore, there is a lack of experimental data collected in a single database for use.  However, models have been developed for extensive vegetation such as woodland for particular frequency ranges and different types of geometry.  

Begin by getting library path

```python
import lib_path
```

Set up the distances within the vegetation (m), the specific attenuation (dB/m) and the maximum attenuation (dB) using the `linspace` routine from `scipy`

```python
from scipy import linspace

distance = linspace(0., 100.0)

specific_attenuation = 0.39

maximum_attenuation = 34.10
```

Set up the keyword args

```python
 kwargs = {'distance': distance,

           'specific_attenuation': specific_attenuation,

           'maximum_attenuation': maximum_attenuation}
```

Calculate the total attenuation due to vegetation using the `vegetation` routines from `wave_propagation`

```python
from Libs.wave_propagation import vegetation

attenuation = vegetation.attenuation(**kwargs)
```

Display the attenuation due to vegetation using the `matplotlib` routines

```python
from matplotlib import pyplot as plt


# Set the figure size

plt.rcParams["figure.figsize"] = (15, 10)


# Set up the axes

fig, axes1 = plt.subplots()

axes2 = axes1.twinx()



# Display the results

axes1.semilogy(distance, attenuation)



# Set the plot title and labels

axes1.set_title('Vegetation Attenuation', size=14)

axes1.set_xlabel('Distance (m)', size=12)

axes1.set_ylabel('Attenuation (dB)', size=12)



# Set the tick label size

axes1.tick_params(labelsize=12)



# Turn on the grid

axes1.grid(linestyle=':', linewidth=0.5)
```
