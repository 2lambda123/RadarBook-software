# Atmospheric Attenuation
***

Referring to Section 2.9.1, dlectromagnetic waves are attenuated in the atmosphere primarily due to oxygen ($O_{2}$) and water vapor ($H_{2}O$). The first peak in attenuation occurs at approximately $22$ GHz and is due to water vapor. The second peak is located at $63$ GHz due to oxygen.  For frequencies up to $1$ THz, the attenuation due to dry air and water vapor is most accurately evaluated by means of a summation of the individual resonance lines from oxygen and water vapor.  Additional factors are added to this result to account for the non-resonant Debye spectrum of oxygen, pressure-induced nitrogen attenuation, and excess water vapor absorption.  The specific atmospheric attenuation is written as (Equation 2.151)

$$
A_s = A_o + A_w = 0.1820\, f\, \big(N_o(f) + N_w(f) \big) \hspace{0.5in} \text{(dB/km)}
$$

The terms $N_o(f)$ and $N_w(f)$ are the imaginary parts of the complex refractivity and are expressed as (Equation 2.152)

$$
N_o(f) = \sum_i S_i^o\,F_i^o + N_d(f), \hspace{0.2in} N_w(f) = \sum_i S_i^w\,F_i^w
$$

For terrestrial paths, the total path attenuation is then found from (Equation 2.153)

$$
A = A_s \, r = (A_o + A_w)\,r \hspace{0.5in} \text{(dB)}
$$

Begin by getting library path

```python
import lib_path
```

Set the start and end frequencies (Hz)

```python
frequency_start = 1

frequency_end = 1e3
```

Set up the frequency array using the `linspace` routine from `scipy`

```python
from scipy import linspace

frequency = linspace(frequency_start, frequency_end, 1000)
```

Set the temperature (K), the dry air pressure (hPa), and the water vapor density (g/m^3)

```python
temperature = 290

dry_air_pressure = 1013.25

water_vapor_density = 7.5
```

Set up the keyword args

```python
kwargs = {'frequency': frequency,

          'temperature': temperature,

          'dry_air_pressure': dry_air_pressure,

          'water_vapor_density': water_vapor_density}
```

Calculate the attenuation using the `attenuation` routine from `atmosphere`

```python
from Libs.wave_propagation import atmosphere

gamma = atmosphere.attenuation(**kwargs)
```

Use the routines from `matplotlib` to display the atmospheric attenuation

```python
from matplotlib import pyplot as plt


# Set the figure size

plt.rcParams["figure.figsize"] = (15, 10)



# Display the results

plt.semilogy(frequency, gamma)



# Set the plot title and labels

plt.title('Atmospheric Attenuation', size=14)

plt.xlabel('Frequency (GHz)', size=12)

plt.ylabel('Specific Attenuation (dB/km)', size=12)



# Set the tick label size

plt.tick_params(labelsize=12)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)
```
