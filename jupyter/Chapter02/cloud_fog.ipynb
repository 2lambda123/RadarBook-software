# Cloud/Fog Attenuation
***

Referring to Section 2.9.4, the attenuation of electromagnetic waves due to clouds and fog is of particular importance not only to airborne and ground-based upward-looking radars, but also to vehicle based radars dealing with automatic emergency braking, radar systems for mining, agriculture, and search and rescue.  Many of these sensors operate in higher frequency regimes where the attenuation due to fog and cloud becomes significant. Considering clouds and fog made up of small droplets less than $0.01$ cm, the Rayleigh approximation allows the attenuation within the cloud or fog to be written as (Equation 2.163)

$$
\gamma = k_l(f, T) \, M \hspace{0.5in} \text{(dB/km)}
$$


This expression is valid for frequencies up to about $200$ GHz.  Typical values for the liquid water density in fog is $0.05$ g/m^{3} for medium fog with a visibility of about $300$ meters, and $0.5$ g/m^{3} for heavy fog with a visibility of about $50$ meters. 
***

Begin by getting library path

```python
import lib_path
```

Set the start and end frequencies (GHz)

```python
frequency_start = 1

frequency_end = 1000
```

Set up the frequency array using the `linspace` routine from `scipy`

```python
from numpy import linspace

frequency = linspace(frequency_start, frequency_end, 2000)
```

Set the liquid water temperature (K) and the liquid water density (g/m^3)

```python
liquid_water_temperature = 290

liquid_water_density = 0.5
```

Set up the keyword args

```python
kwargs = {'frequency': frequency,

          'liquid_water_temperature': liquid_water_temperature,

          'liquid_water_density': liquid_water_density}
```

Calculate the attenuation using the `attenuation` routine from `cloud_fog`

```python
from Libs.wave_propagation import cloud_fog

gamma = cloud_fog.attenuation(**kwargs)
```

Display the cloud/fog attenuation using the routines from `matplotlib`

```python
from matplotlib import pyplot as plt


# Set the figure size

plt.rcParams["figure.figsize"] = (15, 10)



# Display the results

plt.loglog(frequency, gamma)
    
    

# Set the plot title and labels

plt.title('Cloud or Fog Attenuation', size=14)

plt.xlabel('Frequency (GHz)', size=12)

plt.ylabel('Specific Attenuation (dB/km)', size=12)



# Set the tick label size

plt.tick_params(labelsize=12)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)
```
