# Frustum Radar Cross Section

Begin by getting the library path

```python
import libs_path
```

Set the operating frequency (Hz), the nose radius (m), the base radisu (m) and the length (m)

```python
frequency = 1e9

nose_radius = 0.1

base_radius = 0.8

length = 3.0
```

Set the incident angles using the `linspace` routine from `scipy`

```python
from scipy import linspace

from scipy.constants import pi

incident_angle = linspace(0, pi, 1801)
```

Calculate the radar cross section (m^2) for the frustum

```python
from Libs.rcs.frustum import radar_cross_section

from scipy import array


rcs = array([radar_cross_section(frequency, nose_radius, base_radius, length, ia) for ia in incident_angle])
```

```python
from matplotlib import pyplot as plt

from scipy import log10, degrees


# Set the figure size

plt.rcParams["figure.figsize"] = (15, 10)



# Display the results

plt.plot(degrees(incident_angle), 10 * log10(rcs), '')



# Set the plot title and labels

plt.title('RCS vs Incident Angle', size=14)

plt.ylabel('RCS (dBsm)', size=12)

plt.xlabel('Incident Angle (deg)', size=12)



# Set the tick label size

plt.tick_params(labelsize=12)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)
```
