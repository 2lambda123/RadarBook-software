# Frustum Radar Cross Section

Referring to Section 7.4.1.7, the frustum geometry is shown in Figure 7.13. An approximation for the radar cross section due to linearly polarized incident energy is given by (Equation 7.60)

$$
    \sigma = \frac{b\, \lambda}{8\pi\sin\theta_i} \tan^2(\theta_i - \alpha)  \hspace{0.5in} \text{(m^2)},
$$

When the incident angle is normal to the side of the frustum, $\theta_i = 90 + \alpha$, an approximation for the radar cross section is (Equation 7.61)

$$
    \sigma = \frac{8\pi\big(z_2^{1.5} - z_1^{1.5} \big)^2 \sin\alpha}{9\lambda\cos^4\alpha} \hspace{0.5in} \text{(m^2)}.
$$

When the incident angle is either $0^o$ or $180^o$, the radar cross section is approximated by a flat circular plate, as given in Table 7.1.

Set the operating frequency (Hz), the nose radius (m), the base radisu (m) and the length (m)

```python
frequency = 1e9

nose_radius = 0.1

base_radius = 0.8

length = 3.0
```

Set the incident angles using the `linspace` routine from `scipy`

```python
from scipy import linspace

from scipy.constants import pi

incident_angle = linspace(0, pi, 1801)
```

Calculate the radar cross section (m^2) for the frustum

```python
from Libs.rcs.frustum import radar_cross_section

from scipy import array


rcs = array([radar_cross_section(frequency, nose_radius, base_radius, length, ia) for ia in incident_angle])
```

```python
from matplotlib import pyplot as plt

from scipy import log10, degrees


# Set the figure size

plt.rcParams["figure.figsize"] = (15, 10)



# Display the results

plt.plot(degrees(incident_angle), 10 * log10(rcs), '')



# Set the plot title and labels

plt.title('RCS vs Incident Angle', size=14)

plt.ylabel('RCS (dBsm)', size=12)

plt.xlabel('Incident Angle (deg)', size=12)



# Set the tick label size

plt.tick_params(labelsize=12)



# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)
```
