# Crossover Range

For an escort type jammer, the crossover range may be expressed as (Equation 11.18)

\begin{equation}\label{eq:crossover_escort}
    r_c = \left[ \frac{P_t\, G_t^2 \,\sigma\, r_j^2 \,B_j}{{ERP}\, \, G_r \,4 \,\pi \,B_r \,L_r}\right]^{1/4} \hspace{0.5in} \text{(m)},
\end{equation}

and for a self-screening type jammer (Equation 11.19)

\begin{equation}\label{eq:crossover_self}
    r_c = \sqrt{\frac{P_t\, G_t^2 \,\sigma\, B_j}{{ERP}\, \, G_r \,4 \,\pi \,B_r \,L_r}} \hspace{0.5in} \text{(m)}
\end{equation}

Begin by getting the library path

```python
import lib_path
```

Set up the jammer effective radiated power (dBW)

```python
jammer_erp = [10, 40]
```

Import the `linspace` routine from `scipy` to create the jammer ERP array

```python
from scipy import linspace

jammer_erp_vector = linspace(float(jammer_erp[0]), float(jammer_erp[1]), 1000)
```

Set the peak power (W), antenna gain (dB), target RCS (dBsm), jammer bandwidth (Hz), jammer ERP (dBW), radar bandwidth (Hz), and losses (dB) for a self-screening type jammer

```python
peak_power = 50e3

antenna_gain = 40

target_rcs = 3.0

jammer_bandwidth = 20e6

radar_bandwidth = 15e6

losses = 9.0
```

Set up the keyword args

```python
kwargs = {'peak_power': peak_power,

          'antenna_gain': 10 ** (antenna_gain / 10.0),
          
          'target_rcs': 10 ** (target_rcs / 10.0),

          'jammer_bandwidth': jammer_bandwidth,

          'effective_radiated_power': 10 ** (jammer_erp_vector / 10.0),

          'radar_bandwidth': radar_bandwidth,

          'losses': 10 ** (losses / 10.0)}
```

Import the `crossover_range_selfscreen` routine from countermeasures

```python
from Libs.ecm.countermeasures import crossover_range_selfscreen
```

Calculate the crossover range for a self-screening type jammer (m)

```python
crossover_range = crossover_range_selfscreen(**kwargs)
```

Import the `matplotlib` routines for plotting the crossover range for a self-screening type jammer

```python
from matplotlib import pyplot as plt

# Set the figure size

plt.rcParams["figure.figsize"] = (15, 10)


# Display the results

plt.plot(jammer_erp_vector, crossover_range, '')


# Set the plot title and labels

plt.title('Crossover Range', size=14)

plt.xlabel('Jammer ERP (dBW)', size=12)

plt.ylabel('Crossover Range (m)', size=12)


# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)


# Set the tick label size

plt.tick_params(labelsize=12)
```
