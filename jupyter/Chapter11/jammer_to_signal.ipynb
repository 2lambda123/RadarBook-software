# Jammer to Signal Ratio
***

The jammer to signal ratio for either escort or self-screening type jammers may be written as (Equation 11.17)

$$
    \frac{J}{S} = \frac{{ERP}\, \,  G_r\, 4 \,\pi \,r_{t}^4 \,B_r \,L_r}{P_t \,G_{t}^2 \,\sigma\, r_j^2\, B_j}.
$$
***

Begin by getting the library path

```python
import lib_path
```

Set the target range (m)

```python
target_range = [10e3, 50e3]
```

Use the `linspace` routine from `scipy` to set up the target range array

```python
from scipy import linspace

target_range_vector = linspace(float(target_range[0]), float(target_range[1]), 1000)
```

Set the peak power (W), antenna gain (dB), target RCS (dBsm), jammer range (m), jammer bandwidth (Hz), ERP (dBW), target_range (m), radar bandwidth (Hz), losses (dB), and antenna gain in the direction of the jammer (dB)

```python
peak_power = 1e6

antenna_gain = 23.0

target_rcs = -10.0

jammer_bandwidth = 50e6

effective_radiated_power = 5.0

radar_bandwidth = 25e6

losses = 2.0

antenna_gain_jammer_direction = 5.0
```

Set up the keyword args

```python
kwargs = {'peak_power': peak_power,

          'antenna_gain': 10 ** (antenna_gain / 10.0),

          'target_rcs': 10 ** (target_rcs / 10.0),

          'jammer_range': target_range_vector,

          'jammer_bandwidth': jammer_bandwidth,

          'effective_radiated_power': 10 ** (effective_radiated_power / 10.0),

          'target_range': target_range_vector,

          'radar_bandwidth': radar_bandwidth,

          'losses': 10 ** (losses / 10.0),

          'antenna_gain_jammer_direction': 10 ** (antenna_gain_jammer_direction / 10.0)}
```

Calculate the jammer to signal ratio using the `jammer_to_signal` routine from `countermeasures`

```python
from Libs.ecm.countermeasures import jammer_to_signal

jsr = jammer_to_signal(**kwargs)
```

Import the `matplotlib` routines for displaying the jammer to signal ration for a self-screening jammer

```python
from matplotlib import pyplot as plt
```

Import the `log10` routine from `scipy` for displaying the jammer to signal ratio in deciBels

```python
from scipy import log10
```

```python
# Set the figure size

plt.rcParams["figure.figsize"] = (15, 10) 
    
plt.plot(target_range_vector / 1e3, 10 * log10(jsr), '')


# Set the plot title and labels

plt.title('Jammer to Signal Ratio', size=14)

plt.xlabel('Target Range (km)', size=12)

plt.ylabel('Jammer to Signal Ratio (dB)', size=12)


# Turn on the grid

plt.grid(linestyle=':', linewidth=0.5)


# Set the tick label size

plt.tick_params(labelsize=12)
```
